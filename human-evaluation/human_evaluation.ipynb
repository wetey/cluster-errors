```python
# load methods
import pandas as pd
from evaluation_utils import get_files
from evaluation_utils import clustering_task
from evaluation_utils import display_clustering_agreement
from evaluation_utils import get_clustering_agreement
from evaluation_utils import average_agreement
```

```python
def pipeline(original_dataset, evaluation_results_path, original_questions_path):
    '''
        input: original dataset: JSON file saved after kmeans clustering
               evaluation_results_path: survey results 
               original_questions_path: evaluation questions generated
        output: survey results and annotator agreement
    '''
    ocq, cq = get_files(evaluation_results_path, original_questions_path)
    original_dataset = pd.read_json(original_dataset)
    
    print('clustering task accruacy: ')
    clustering_task(ocq, cq)

    print('clustering annotator agreement: ')
    display_clustering_agreement(cq)

    c_agreement = get_clustering_agreement(cq)

    c_average_agreement, c_sum_agreement, c_count = average_agreement(c_agreement)
    print(f'average agreeement for clustering = {c_average_agreement}')

    return ocq, cq, c_sum_agreement, c_count
```
